```python
import math
strInput = input("Do you want to use advanced or basic calculations? (Y/N) ")
if strInput == "Y":
    ad_op = input("Would you like to do cubing, squaring, and square roots? (C/S/SR) ")
    ad_num1 = float(input("Enter number: "))
    if ad_op == "C":
        print(ad_num1 * ad_num1 * ad_num1)
    elif ad_op == "S":
        print(ad_num1 * ad_num1)
    elif ad_op == "SR":
        print(math.sqrt(ad_num1))
    else:
        print("Invalid answer")
        
elif strInput == "N":
    op = input("Would you like to do adding, subtracting, multiplication, or division? (+ or - or * or /) ")
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    if op == "+":
        print(num1 + num2)
    elif op == "-":
        print(num1 - num2)
    elif op == "/":
        print(num1 / num2)
    elif op == "*":
        print(num1 * num2)
    else:
        print("Invalid answer")

else:
    print("Invalid answer")
```

```python
def translate(phrase):
    translation = ""
    for letter in phrase:
        if  letter.lower() in "aeiou":
            if letter.isupper():
                translation = translation + "G"
            else:
                translation = translation + "g"
        else:
            translation = translation + letter
    return translation
print(translate(input("Enter a phrase: ")))
```

```python
question_prompts = [
    "What color are apples?\n(a) Red/Green\n(b) Purple\n(c) Orange\n\n",
    "What color are bananas?\n(a) Teal\n(b) Magenta\n(c) Yellow\n\n",
    "What color are strawberries?\n(a) Yellow\n(b) Red\n(c) Blue\n\n"
]
class Question:
    def __init__(self, prompt, answer):
        self.prompt = prompt
        self.answer = answer
questions = [
    Question(question_prompts[0], "a"),
    Question(question_prompts[1], "c"),
    Question(question_prompts[2], "b")
]
def run_test(questions):
    score = 0
    for question in questions:
        answer = input(question.prompt)
        if answer == question.answer:
            score += 1
    print("You got " + str(score) + "/" + str(len(questions)) + " correct")
run_test(questions)
```
